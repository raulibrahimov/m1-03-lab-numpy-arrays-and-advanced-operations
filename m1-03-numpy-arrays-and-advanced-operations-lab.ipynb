<h1>TASK-1</h1>

```python
import numpy as np
```

```python
rng = np.random.default_rng(seed=0)
```

```python
readings = rng.normal(50, 10, size=(360,4))

print(readings.shape)

print(readings.dtype)

print(readings[:3])
```

<h2>TASK-2</h2>

```python
scaled_readings = (readings -  readings.min())/(readings.max() -  readings.min())
scaled_readings
```

```python
scaled_readings.max()
```

```python
scaled_readings.min()
```

```python
scaled_readings.mean(axis=0)
```

```python
centered_readings = readings  -  readings.mean(axis=0)

centered_readings
```

<h1>TASK-3</h1>

```python
outlier_mask = (readings > 80) | (readings < 20)


outlier_mask.sum()

(outlier_mask.sum()/readings.size)*100
```

```python
cleaned_readings = readings.astype(float).copy()

cleaned_readings[outlier_mask] = np.nan
cleaned_readings
```

```python
np.isnan(cleaned_readings).sum()
```

<h1>TASK-4</h1>

```python
sensor_means = np.array(readings.mean(axis=0))
sensor_stds = np.array(readings.std(axis=0))

time_means = np.array(readings.mean(axis=1))
time_means.shape
sensor_means.shape
```

<h1>TASK-5</h1>

```python


report = {"total_readings" : readings.size, "outlier_count" : int(np.isnan(cleaned_readings).sum()), "outlier_percent" : (int(np.isnan(cleaned_readings).sum())/outlier_mask.size)*100 ,"sensor_means" : sensor_means , "sensor_stds" :sensor_stds}


#Report

for key,value in report.items():
    print(key,value)

report["total_readings"] == readings.size
```

<h1>PART-2</h1>
<h1>TASK-6</h1>

```python
rng = np.random.default_rng(seed=0)

sim = np.array(rng.normal(0,1,size=(1000,6)))

print(sim.shape)


print(sim.dtype)

print(sim[:3])
```

<h1>TASK-7</h1>

```python
scenario_shift = np.array([-0.3, -0.1, 0.0, 0.1, 0.2, 0.4])


adjusted_sim = sim + scenario_shift

adjusted_sim.shape == sim.shape

adjusted_sim_mean = adjusted_sim.mean(axis=0)

sim_means = sim.mean(axis=0)

print(sim_means - adjusted_sim_mean)
```

<h1>TASK-8</h1>

```python
scenario_means = np.array(adjusted_sim.mean(axis=0))

np.argsort(scenario_means)

top_2 = np.partition(scenario_means,-2)


top_2[-2:]
```

<h1>TASK-9</h1>

```python
rng_2 = np.random.default_rng(seed=0)

sim_2 = np.array(rng_2.normal(0,1,size=(1000,6)))

rng_3 = np.random.default_rng(seed=1)

sim_3 = np.array(rng_3.normal(0,1,size=(1000,6)))
sim_2.shape == sim_3.shape

np.allclose(sim_2, sim_3)

top_two_values = top_2[-2:]

c_scenario_means = scenario_means.copy()

c_scenario_means = c_scenario_means.tolist()

top_two_indices = [c_scenario_means.index(value) for value in top_two_values]

scenario_means[top_two_indices]
```

<h1>TASK-10</h1>

```python


report = {
    "shape": adjusted_sim.shape,
    "scenario_means": scenario_means,
    "top_two_indices": top_two_indices,
    "top_two_values": top_two_values.tolist()
}

assert len(scenario_means) == 6, "scenario_means should have length 6"
assert len(top_two_indices) == 2, "top_two_indices should have length 2"


report_string = (
    f"Simulation Shape: {report['shape']}\n"
    f"Scenario Means: {report['scenario_means']}\n"
    f"Top Two Scenario Indices: {report['top_two_indices']}\n"
    f"Top Two Scenario Values: {report['top_two_values']}"
)

print(report_string)
```
